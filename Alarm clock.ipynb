```python
import time
import datetime
import os

def set_alarm(alarm_time):
    while True:
        # Get the current time
        current_time = datetime.datetime.now().strftime("%H:%M:%S")
        print("Current time:", current_time)

        # Check if the current time matches the alarm time
        if current_time == alarm_time:
            print("Wake up! It's time!")

            # Play an alarm sound (You can change this to a path of a sound file on your system)
            os.system('say "Time to wake up!"')  # For MacOS, use `afplay /path/to/soundfile.wav`
            break

        # Sleep for 1 second before checking the time again
        # time.sleep(1)

# Set the alarm time in HH:MM:SS format
alarm_time = input("Enter the time to set the alarm (HH:MM:SS): ")
set_alarm(alarm_time)


```
